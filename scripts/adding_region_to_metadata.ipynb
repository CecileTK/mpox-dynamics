```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../monkeypox-build/data/nextstrain__metadata_1000.tsv", sep = "\t")
```

```python
df
```

```python
north_america = ['Canada','USA']
western_europe = ['Austria','Germany','Switzerland','United Kingdom','France','Belgium']
central_europe = ['Slovakia','Slovenia']
southern_europe = ['Italy','Portugal','Spain']
south_american = ['Colombia','Peru']
```

```python
df["region_coarse"] = np.nan
df.region_coarse[df.country.isin(north_america)] = "North America"
df['region_coarse'][df.country.isin(western_europe)] = "Western Europe"
df['region_coarse'][df.country.isin(central_europe)] = "Central Europe"
df['region_coarse'][df.country.isin(southern_europe)] = "Southern Europe"
df['region_coarse'][df.country.isin(south_american)] = "South America"
```

```python
df = df[["accession", "date", "region_coarse", "country", "strain"]]
df
```

```python
df.to_csv("../monkeypox-build/data/metadata_region_1000.tsv", sep = "\t")
```
