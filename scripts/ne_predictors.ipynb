```python
import pandas as pd
import numpy as np
import altair as alt
alt.data_transformers.disable_max_rows()
```

```python
df = pd.read_csv("/Users/mparedes/Downloads/owid-monkeypox-data.csv") # from https://ourworldindata.org/explorers/monkeypox
```

```python
df
```

```python
df.location = df.location.astype("str")
```

```python
#the only countries with enough sequences (>5)
countries = ["Austria",
"Belgium",
"Canada",
"Colombia",
"France",
"Germany",
"Italy",
"Peru",
"Portugal",
"Slovakia",
"Slovenia",
"Spain",
"Switzerland",
"United States",
"United Kingdom",
]

regions = ["NorthAmerica", "WesternEurope", "CentralEurope", "SouthernEurope", "SouthAmerica"]

north_america = ['Canada','United States']
western_europe = ['Austria','Germany','Switzerland','United Kingdom','France','Belgium']
central_europe = ['Slovakia','Slovenia']
southern_europe = ['Italy','Portugal','Spain']
south_american = ['Colombia','Peru']
```

```python
df["region_coarse"] = np.nan
df["region_coarse"][df.location.isin(north_america)] = "NorthAmerica"
df['region_coarse'][df.location.isin(western_europe)] = "WesternEurope"
df['region_coarse'][df.location.isin(central_europe)] = "CentralEurope"
df['region_coarse'][df.location.isin(southern_europe)] = "SouthernEurope"
df['region_coarse'][df.location.isin(south_american)] = "SouthAmerica"
```

```python
df.region_coarse.value_counts()
```

```python
df_country = df[df.location.isin(countries)]
df_plot_country = df_country[["location", "date", "new_cases_smoothed_per_million"]]
df_country = df_country[["location", "date", "new_cases_smoothed"]]
df_country
```

```python
hosp_plot =  alt.Chart(df_plot_country, width = 750).mark_line(interpolate='monotone', opacity = 1).encode(
    x=alt.X('date:T', axis=alt.Axis(title=None, grid=False, format="%B %Y")),
    y=alt.Y('new_cases_smoothed_per_million:Q',axis=alt.Axis(title = "", grid=False)),
    color = alt.Color("location:N")).properties(
    width=900,
    height=500
)
hosp_plot
```

```python
#wide format for easy beauti input
wide_df = pd.pivot(df_country, index="location", columns="date", values="new_cases_smoothed")
wide_df = wide_df.replace(np.nan, 0.001)
wide_df = wide_df.replace(0.00, 0.001) #can't have 0 values for log standarization
wide_df
```

```python
rev_wide = wide_df[wide_df.columns[::-1]]
```

```python
x = list(range(0,231,1))
rev_wide.columns = x
```

```python
rev_wide = rev_wide.rename({"United States":"USA","United Kingdom":"UnitedKingdom"}, axis='rows')
```

```python
rev_wide
```

```python
rev_wide.to_csv("../mascot_glm/data/ne_case_pred.csv")
```

```python
country_plot =  alt.Chart(df_country, width = 750).mark_line(interpolate='monotone', opacity = 1).encode(
    x=alt.X('date:T', axis=alt.Axis(title=None, grid=False, format="%B %Y")),
    y=alt.Y('new_cases_smoothed:Q',axis=alt.Axis(title = " Daily Mpox Cases (smoothed)", grid=False),  stack = None,scale=alt.Scale(domain=(0, 17))),
    color = alt.Color("location:N", legend=alt.Legend(orient = "left",offset = - 180, labelFontSize = 14, titleFontSize = 14, symbolSize = 110))).properties(
    width=1050,
    height=200
)
country_plot
```

```python
#ne predictors for region level analysis
```

```python
df_region = df[df.region_coarse.isin(regions)]
df_region_plot = df_region[["region_coarse", "date", "new_cases_smoothed_per_million"]]
df_region = df_region[["region_coarse", "date", "new_cases_smoothed"]]
df_region
```

```python
df_region_plot = df_region_plot.groupby(['region_coarse', pd.Grouper(key='date')])['new_cases_smoothed_per_million'].sum()
df_region_plot = df_region_plot.reset_index()


hosp_plot =  alt.Chart(df_region_plot, width = 750).mark_line(interpolate='monotone', opacity = 1).encode(
    x=alt.X('date:T', axis=alt.Axis(title=None, grid=False, format="%B %Y")),
    y=alt.Y('new_cases_smoothed_per_million:Q',axis=alt.Axis(title = "", grid=False)),
    color = alt.Color("region_coarse:N")).properties(
    width=900,
    height=500
)
hosp_plot
```

```python
df_region = df_region.groupby(['region_coarse', pd.Grouper(key='date')])['new_cases_smoothed'].sum()
df_region = df_region.reset_index()
df_region
```

```python
#wide format for easy beauti input
w_region_df = pd.pivot(df_region, index="region_coarse", columns="date", values="new_cases_smoothed")
w_region_df = w_region_df.replace(np.nan, 0.001)
w_region_df = w_region_df.replace(0.00, 0.001) #can't have 0 values for log standarization
w_region_df
```

```python
rev_region_wide = w_region_df[wide_df.columns[::-1]]
x = list(range(0,231,1))
rev_region_wide.columns = x
rev_region_wide = rev_region_wide.rename({"United States":"USA","United Kingdom":"UnitedKingdom"}, axis='rows')
rev_region_wide
```

```python
rev_region_wide.to_csv("../mascot_glm/data/ne_case_pred_regions.csv")
```

```python
region_plot =  alt.Chart(df_region, width = 750).mark_line(interpolate='monotone', opacity = 1).encode(
    x=alt.X('date:T', axis=alt.Axis(title=None, grid=False, format="%B %Y")),
    y=alt.Y('new_cases_smoothed:Q',axis=alt.Axis(title = " Daily Mpox Cases (smoothed)", grid=False),  stack = None,scale=alt.Scale(domain=(0, 17))),
    color = alt.Color("region_coarse:N", legend=alt.Legend(orient = "left",offset = - 180, labelFontSize = 14, titleFontSize = 14, symbolSize = 110))).properties(
    width=1050,
    height=200
)
region_plot
```
