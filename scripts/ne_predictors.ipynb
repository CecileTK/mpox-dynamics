```python
#the only countries with enough sequences (>5)
countries = ["Austria",
"Belgium",
"Canada",
"Colombia",
"France",
"Germany",
"Italy",
"Peru",
"Portugal",
"Slovakia",
"Slovenia",
"Spain",
"Switzerland",
"United States",
"United Kingdom",
]
```

```python
df = df[df.location.isin(countries)]
df = df[["location", "date", "new_cases_smoothed"]]
df
```

```python
#wide format for easy beauti input
wide_df = pd.pivot(df, index="location", columns="date", values="new_cases_smoothed")
wide_df = wide_df.replace(np.nan, 0.001)
wide_df = wide_df.replace(0.00, 0.001) #can't have 0 values for log standarization
wide_df
```

```python
rev_wide = wide_df[wide_df.columns[::-1]]
```

```python
x = list(range(0,231,1))
rev_wide.columns = x
```

```python
rev_wide = rev_wide.rename({"United States":"USA","United Kingdom":"UnitedKingdom"}, axis='rows')
```
