```python

def reindex_by_date(func_df):
    func_df = func_df.reindex(pd.bdate_range('2022-04-01', '2023-01-30', freq = "W-Sat"))
    func_df['region']=func_df['region'].bfill()
    if "median_prev" in func_df.columns:
        #print("pass")
        func_df["median_prev"] = func_df["median_prev"].bfill().ffill()
    return (func_df)
```

```python
## now we do prevalence
```

```python
case_prevalence = pd.read_csv("../case-rt-analysis/estimates/case-prevalence-estimates_region.tsv", sep="\t", parse_dates= ["date"])
```

```python
case_prevalence
```

```python
case_prevalence = case_prevalence[["region", "date", "median_prev"]]
case_prevalence = case_prevalence.groupby(['region', pd.Grouper(key='date')])['median_prev'].sum()
case_prevalence = case_prevalence.reset_index()
case_prevalence
```

```python
#resample to go from daily to weekly
case_prevalence.index = case_prevalence.date.astype('datetime64[ns]')
grouper = case_prevalence.groupby([pd.Grouper(freq='W-Sat'), 'region'])
result = grouper['median_prev'].sum()
weekly_region_prev = pd.DataFrame(result).reset_index()
weekly_region_prev.index = weekly_region_prev.date.astype('datetime64[ns]')

weekly_region_prev_bfill= weekly_region_prev.groupby('region').apply(reindex_by_date).drop(columns =["date", "region"]).reset_index().rename(columns = {"level_1": "date" })

w_region_prev_df = pd.pivot(weekly_region_prev_bfill, index="region", columns="date", values="median_prev")
w_region_prev_df = w_region_prev_df.replace(np.nan, 0.1)
w_region_prev_df = w_region_prev_df.replace(0.00, 0.1) #can't have 0 values for log standarization
w_region_prev_df

rev_region_wide = w_region_prev_df[w_region_prev_df.columns[::-1]]
x = list(range(0,44,1))
rev_region_wide.columns = x
rev_region_wide = rev_region_wide.rename({"North America": "NorthAmerica", "Western Europe": "WesternEurope", "Central Europe":"CentralEurope", "South America":"SouthAmerica", "Southern Europe": "SouthernEurope"}, axis='rows')

rev_region_wide
```
