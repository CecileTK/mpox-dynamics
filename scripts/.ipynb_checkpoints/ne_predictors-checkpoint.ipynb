```python
month_pred = weekly_region_prev[["date", "region"]]
```

```python
test = month_pred
#test.index = test.date
test
```

```python

def reindex_by_date(df):
    df = df.reindex(pd.bdate_range('2022-04-01', '2023-01-30', freq = "W-Sat"))
    df['region']=df['region'].bfill()
    if "median_prev" in df.columns:
        df["median_prev"] = df['median_prev'].bfill().ffill()
    return (df)

test_new= test.groupby('region').apply(reindex_by_date).drop(columns =["date", "region"]).reset_index().rename(columns = {"level_1": "date" })
test_new
```

```python
test_new["month"] = pd.to_datetime(test_new['date']).dt.to_period('M')
```

```python

for x in range(0,10,1):
    test_new["month_"+str(x)] = 0
    test_new["month_"+ str(x)][test_new.month == ((pd.to_datetime("2022-04") + relativedelta(months=x)).to_period('M'))] = 1
```

```python
month_pred = test_new
month_pred
```

```python
df_list = []
for i in range(0,10,1):
    df_name = ("foo" + "_" + str(i))
    #print(df_list)
    value = ("month_"+ str(i))
    df_name = pd.pivot(month_pred, index="region", columns="date", values=value)

    df_name = df_name.replace(np.nan, 0.0)
    df_name = df_name.replace(0.00, 0.0) #can't have 0 values for log standarization

    df_name = df_name[df_name.columns[::-1]]
    x = list(range(0,44,1))
    df_name.columns = x
    df_name = df_name.rename({"North America": "NorthAmerica", "Western Europe": "WesternEurope", "Central Europe":"CentralEurope", "South America":"SouthAmerica", "Southern Europe": "SouthernEurope"}, axis='rows')
    df_list.append(df_name)
```
