```python
import pandas as pd
import numpy as np
import altair as alt
import matplotlib as plt
```

```python
metadata = pd.read_csv("../monkeypox-build/data/nextstrain__metadata_1000.tsv", sep = "\t")
```

```python
meta_b = metadata[metadata.lineage.str.contains('^B.*')==True]
```

```python
meta_b = meta_b[meta_b['date'].notna()]
meta_b["year_month"] = meta_b['date'].str.rsplit("-", 1).str[-0].tolist()
meta_b = meta_b[meta_b.year_month.str.contains('XX*')==False]
meta_b
```

```python
country_list = meta_b.country.value_counts().loc[lambda x: x>5].index.tolist()
meta_b = meta_b[meta_b.country.isin(country_list)]
meta_b = meta_b[meta_b.country != "Brazil"]
meta_b
```

```python
meta_b[meta_b.country == "Portugal"].year_month.value_counts()
```

```python
country_count = meta_b.groupby(["year_month"])["country"].value_counts().reset_index(name="count")
```

```python
range_ = ["#5E1D9D","#492AB5","#4042C7","#3E5DD0","#4377CD","#4A8CC2","#549DB2","#60AA9E","#6EB389","#80B974","#92BC63","#A6BE55","#B9BC4A","#CBB742","#D9AD3D","#E29D39","#E68634","#E56A2F","#E04929","#DB2823"]
alt.Chart(country_count).mark_bar().encode(
    alt.X("year_month",axis=alt.Axis( grid=False, )),
    alt.Y("count",axis=alt.Axis( grid=False)), 
    alt.Color("country:N", scale=alt.Scale(scheme="category20"))).properties(
    width = 800,
    height = 600
)
```

```python
# Getting data
raw_cases = pd.read_csv("../case-rt-analysis/data/monkeypox-cases-counts.tsv", sep="\t")
raw_cases.location.value_counts()
```

```python
raw_cases= raw_cases[raw_cases.location != "World"]
raw_cases.location[raw_cases.location == "United States"] = "USA"
```

```python
raw_cases[(raw_cases.location == "United Kingdom") & (raw_cases.date > "2022-10-01")]
```

```python
raw_cases.set_index('date', inplace=True)
raw_cases.index = pd.to_datetime(raw_cases.index)
weekly_prevalence = raw_cases.groupby(["location"]).resample('1W').sum().reset_index()
#month_prevalence["year-month"]= pd.to_datetime(month_prevalence['date']).dt.to_period('M')
weekly_prevalence
```

```python
alt.Chart(weekly_prevalence).mark_line(interpolate = "monotone").encode(
    alt.X("date:T", axis=alt.Axis( grid=False)),
    alt.Y("cases"), 
    alt.Color("location:N", scale=alt.Scale(scheme="category20"))).properties(
    width = 800,
    height = 500
)
```

```python
## now read in prevalence estimates from Marlin's evofr models
case_incidence = pd.read_csv("../case-rt-analysis/estimates/case-incidence-estimates_country.tsv", sep="\t", parse_dates= ["date"])
```

```python
case_incidence
```

```python
case_incidence.location[case_incidence.location == "United States"] = "USA"
case_incidence.set_index('date', inplace=True)
case_incidence.index = pd.to_datetime(case_incidence.index)
weekly_incidence = case_incidence.groupby(["location"]).resample('1W').sum().reset_index()
#month_prevalence["year-month"]= pd.to_datetime(month_prevalence['date']).dt.to_period('M')
weekly_incidence
```

```python
alt.Chart(weekly_incidence).mark_line(interpolate = "monotone").encode(
    alt.X("date:T", axis=alt.Axis( grid=False, format="%B %Y")),
    alt.Y("median_I_smooth"), 
    alt.Color("location:N", scale=alt.Scale(scheme="category20"))).properties(
    width = 800,
    height = 500
)
```
