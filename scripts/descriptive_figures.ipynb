```python
import pandas as pd
import numpy as np
import altair as alt
import matplotlib as plt
import altair_saver
from altair import datum
```

```python
metadata = pd.read_csv("../monkeypox-build/data/nextstrain__metadata_3000.tsv", sep = "\t")
```

```python
meta_b = metadata[metadata.lineage.str.contains('^B.*')==True]
meta_b = meta_b[meta_b['date'].notna()]
meta_b["year_month"] = meta_b['date'].str.rsplit("-", 1).str[-0].tolist()
meta_b = meta_b[meta_b.year_month.str.contains('XX*')==False]
meta_b
```

```python
region = {'Canada':'North America',
          'USA':'North America',
         'Finland':'Northern Europe',
           'Austria': 'Western Europe',
        'Germany':'Western Europe', 
        'Switzerland':'Western Europe',
         'United Kingdom':'Western Europe',
        'France':'Western Europe',
        'Belgium':'Western Europe',
        'Netherlands':'Western Europe',
        'Slovakia':'Central Europe',
        'Slovenia':'Central Europe',
        'Italy':'Southern Europe',
        'Portugal':'Southern Europe',
        'Spain':'Southern Europe',
         'Colombia':'South America',
          
        'Peru':'South America'}

country_list = meta_b.country.value_counts().loc[lambda x: x>5].index.tolist()
meta_b = meta_b[meta_b.country.isin(country_list)]
meta_b = meta_b[meta_b.country != "Brazil"]
meta_b = meta_b[meta_b.country != "Netherlands"]
meta_b["region"] = np.nan
meta_b.region = meta_b.country.map(region)
#meta_b = meta_b.dropna()

meta_b
```

```python

meta_b.date = pd.to_datetime(meta_b.date)
meta_b['year_week'] = meta_b['date'].dt.year.astype(str) + '-' + meta_b['date'].dt.week.astype(str).str.zfill(2)
```

```python
# Getting data
raw_cases = pd.read_csv("../case-rt-analysis/data/monkeypox-cases-counts.tsv", sep="\t")
raw_cases.location.value_counts()
```

```python
raw_cases= raw_cases[raw_cases.location != "World"]
raw_cases.location[raw_cases.location == "United States"] = "USA"
raw_cases[(raw_cases.location == "United Kingdom") & (raw_cases.date > "2022-10-01")]
raw_cases.set_index('date', inplace=True)
raw_cases.index = pd.to_datetime(raw_cases.index)
raw_cases["region"] = np.nan
raw_cases.region = raw_cases.location.map(region)
weekly_prevalence = raw_cases.groupby(["location"]).resample('1W').sum().reset_index()
weekly_count_region = raw_cases.groupby(["region"]).resample('1W').sum().reset_index()

#month_prevalence["year-month"]= pd.to_datetime(month_prevalence['date']).dt.to_period('M')
weekly_prevalence
```

```python
weekly_prevalence['order'] = weekly_prevalence['location'].replace(
    {val: i for i, val in enumerate(colors.keys())}
)

weekly_cases_country = alt.Chart(weekly_prevalence).mark_bar(interpolate = "monotone", size =20).encode(
    alt.X("date:T", axis=alt.Axis( title="", grid=False, format="%B %Y")),
    alt.Y("cases", axis=alt.Axis( grid=False, title= "Weekly Mpox Cases")), 
    alt.Color("location:N", title = "country",  scale=alt.Scale(domain=list(colors.keys()), range=list(colors.values())), legend=alt.Legend(title = "Country", orient = "right",offset = -165, labelFontSize = 20, titleFontSize = 20, symbolSize = 110)), 
    alt.Order("order", sort = "descending")).properties(
    width = 800,
    height = 500
)
weekly_cases_country
```

```python
domain = ['Central Europe', 'North America', 'South America' ,"Southern Europe", "Western Europe"]
range_ = ['#EEC060', '#2664A5', '#A76BB1', "#EEA160", "#356D4C"]
weekly_cases_region = alt.Chart(weekly_count_region).mark_bar(interpolate = "monotone", size =20).encode(
    alt.X("date:T", axis=alt.Axis( title="", grid=False, format="%B %Y")),
    alt.Y("cases", axis=alt.Axis( grid=False, title = "Weekly Mpox Cases")), 
    alt.Color("region:N", scale=alt.Scale(domain = domain, range=range_), legend=alt.Legend(title = "Region", orient = "right",offset = -165, labelFontSize = 20, titleFontSize = 20, symbolSize = 110))).properties(
    width = 800,
    height = 500
)
weekly_cases_region
```

```python
# Add labels to each chart
weekly_cases_country = weekly_cases_country.properties(title='A')
monthly_seqs_country = monthly_seqs_country.properties(title='C')
weekly_cases_region = weekly_cases_region.properties(title='B')
monthly_seqs_region = monthly_seqs_region.properties(title='D')

# Concatenate labeled charts A and B horizontally
chart_ab_labeled = alt.hconcat(weekly_cases_country, weekly_cases_region, spacing=50).resolve_scale(x = "shared", y='independent', color = "independent" )

# Concatenate labeled charts C and D horizontally
chart_cd_labeled = alt.hconcat(monthly_seqs_country, monthly_seqs_region, spacing=50).resolve_scale(x = "shared", y='independent', color = "independent")

# Concatenate labeled charts AB and CD vertically
chart_abcd_labeled = alt.vconcat(chart_ab_labeled, chart_cd_labeled, spacing=50).resolve_scale(x='shared', color ='independent')

chart_abcd_labeled.configure_view(
    strokeWidth=0
).configure_title(
    anchor='start', fontSize= 35
).configure_axis(
    labelFontSize=24,
    titleFontSize=24
).configure_legend(
    labelFontSize = 40)
```
